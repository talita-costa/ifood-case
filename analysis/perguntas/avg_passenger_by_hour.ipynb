#pergunta 2

#Qual a mÃ©dia de passageiros (passenger_count) por cada hora do dia que pegaram tÃ¡xi no mÃªs de maio considerando todos os tÃ¡xis da frota?

#usarei o mÃªs 05/23 e os tÃ¡xis do tipo yellow e green como referÃªncia

```python
#ler os arquivos dos yellow taxi no mÃªs 5 do ano de 2023
df_yellow = pd.read_parquet('s3://talita-costa-pipeline/silver/ny/taxi_rides/yellow/')
df_yellow = df_yellow.query("year == 2023 and month == 5")

df_yellow.info()
display(df_yellow.head(10))
```

```python
#ler os arquivos dos green taxi no mÃªs 5 do ano de 2023
df_green = pd.read_parquet('s3://talita-costa-pipeline/silver/ny/taxi_rides/green/')
df_green = df_green.query("year == 2023 and month == 5")

df_green.info()
display(df_green.head(10))
```

```python
#fazer o join entre o df_yellow e o df_green
df = pd.concat([df_yellow, df_green], ignore_index=True)

df.info()
display(df.head(10))
```

```python
#criar a coluna 'hora' extraindo a hora da coluna 'pickup_datetime'
df['hora'] = df['pickup_datetime'].dt.hour

#criar a coluna 'ano' e 'mes' extraindo das colunas da tabela
ano = df['year'].iloc[0]
mes = df['month'].iloc[0]

#calcular a media dos passageiros (passenger_count) por cada hora do dia no mes de maio de 2023
avg_passenger_by_hour = (
    df.groupby('hora')['passenger_count']
    .mean()
    .reset_index()
    .rename(columns={'passenger_count': 'avg_passenger'})
    .sort_values('hora')
)

#montar o resultado
avg_passenger_by_hour['ano'] = ano
avg_passenger_by_hour['mes'] = mes

df_resultado = avg_passenger_by_hour[['ano', 'mes', 'hora', 'avg_passenger']]

#mostrar o resultado
styled = (
    df_resultado.style
    .format({'avg_passenger': "{:.2f}"})
    .set_properties(**{'background-color': '#f0f0f0', 'color': 'black', 'border-color': 'black'})
    .set_table_styles([
        {'selector': 'th', 'props': [('background-color', '#40466e'), ('color', 'white')]},
        {'selector': 'td', 'props': [('text-align', 'center')]},
    ])
    .set_caption("ðŸš– MÃ©dia de passageiros por hora - Maio/2023")
)

styled
```
