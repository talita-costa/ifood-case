```python
import pandas as pd
```

#pergunta 1

#qual a média de valor total (total_amount) recebido em um mês considerando todos os yellow táxis da frota?

#usarei o mês 05/23 como referência

```python
#ler os arquivos dos yellow taxi no mês 5 do ano de 2023
df = pd.read_parquet('s3://talita-costa-pipeline/silver/ny/taxi_rides/yellow/')
df = df.query("year == 2023 and month == 5")

df.info()
display(df.head(10))
```

```python
#calcular a media do valor total (total_amount) recebido no mês 5 do ano de 2023 dos yellow táxis da frota
ano = df['year'].iloc[0]
mes = df['month'].iloc[0]
avg_total_amount = df['total_amount'].mean()

#montar o resultado
df_resultado = pd.DataFrame({
    'ano': [ano],
    'mes': [mes],
    'avg_total_amount': [round(avg_total_amount, 2)]
})

#mostrar o resultado
styled = (
    df_resultado.style
    .format({'avg_total_amount': "{:.2f}"})
    .set_properties(**{'background-color': '#f0f0f0', 'color': 'black', 'border-color': 'black'})
    .set_table_styles([
        {'selector': 'th', 'props': [('background-color', '#40466e'), ('color', 'white')]},
        {'selector': 'td', 'props': [('text-align', 'center')]},
    ])
    .set_caption("🚖 Média do Total Amount - Yellow Taxi - 2023/05")
)

styled
```
