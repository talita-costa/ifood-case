```python
#O critério de segmentação por cliente que faz sentido para o teste A/B em análise é pelo critério minimum_order_value
#O objetivo é analisar se houve aumento significativo em alguma das faixas de valores com o uso de cupons nos restaurantes ativos
#As faixas de valores são: '0-30', '30-60', '60-100', '100-200', '200+'
```

```python
import pandas as pd
import plotly.express as px
```

```python
#Lendo a tabela order_clean
orders_clean = pd.read_parquet("../data/order_clean.parquet")
display(orders_clean.head())
display(orders_clean.info())

#Lendo a tabela restaurant_clean
restaurant_clean = pd.read_parquet("../data/restaurant_clean.parquet")
display(restaurant_clean.head())
display(restaurant_clean.info())

#Lendo a tabela consumer_merge
consumer_merge = pd.read_parquet("../data/consumer_merge.parquet")
display(consumer_merge.head())
display(consumer_merge.info())
```

```python
#Filtrando apenas pedidos de restaurantes ativos
restaurant_ativos = restaurant_clean[restaurant_clean['enabled'] == True]
```

```python
#Fazendo o merge entre as tabelas order_clean e consumer_merge para trazer a coluna is_target
order_target = orders_clean.merge(consumer_merge[['customer_id', 'is_target']], on='customer_id', how='left')
display(order_target.head())
display(order_target.info())
```

```python
#Fazendo o merge entre as tabelas order_target e restaurant_ativos para trazer a coluna minimum_order_value
order_rest = order_target.merge(restaurant_ativos[['id', 'minimum_order_value']], left_on='merchant_id', right_on='id', how='left')
order_rest = order_rest[order_rest['minimum_order_value'].notnull()]
display(order_rest.head())
display(order_rest.info())
```

```python
#Calculando o ticket medio por customer_id e mês

ticket_medio_consumer = (order_rest.groupby(['customer_id', 'ano_mes_created_at'])['order_total_amount']
    .mean()
    .reset_index()
    .rename(columns={'order_total_amount': 'ticket_medio'})
)

min_order_by_consumer = (order_rest.groupby('customer_id')['minimum_order_value']
    .mean()
    .reset_index()
)

ticket_segmented = ticket_medio_consumer.merge(min_order_by_consumer, on='customer_id', how='left')

ticket_segmented = ticket_segmented.merge(consumer_merge[['customer_id', 'is_target']], on='customer_id', how='left')

ticket_segmented['ano_mes_created_at'] = ticket_segmented['ano_mes_created_at'].dt.strftime('%Y-%m')

test_ab = ['2018-12', '2019-01']
ticket_segmented = ticket_segmented[ticket_segmented['ano_mes_created_at'].isin(test_ab)]

ticket_segmented['faixa_min_order_value'] = pd.cut(ticket_segmented['minimum_order_value'],
    bins=[0, 30, 60, 100, 200, 10000],
    labels=['0-30', '30-60', '60-100', '100-200', '200+']
)

ticket_segmented_avg = (ticket_segmented.groupby(['faixa_min_order_value', 'ano_mes_created_at', 'is_target'])['ticket_medio']
    .mean()
    .reset_index()
)

display(ticket_segmented_avg.head())
display(ticket_segmented_avg.info())
```

```python
#Filtrando apenas o grupo target (grupo de interesse)
ticket_segmented_target = ticket_segmented_avg[ticket_segmented_avg['is_target'] == 'target']

#Resultado por gráfico
fig = px.bar(
    ticket_segmented_target,
    x='faixa_min_order_value',
    y='ticket_medio',
    color='ano_mes_created_at',
    barmode='group',
    labels={
        'faixa_min_order_value': 'Faixa de Minimum Order Value',
        'ticket_medio': 'Ticket Médio',
        'ano_mes_created_at': 'Mês'
    },
    title='Ticket Médio por Faixa de Minimum Order Value (Grupo Target) - Dez/2018 vs Jan/2019'
)

fig.show()
```
